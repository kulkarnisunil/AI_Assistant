### 14] How does the “in” keyword work differently for lists and dictionaries?

Answer:- With lists, in checks for a value using linear search. With dictionaries, in checks for a key using hashing — not the value unless you explicitly check .values().

```python
# List
nums = [10, 20, 30]
print(20 in nums)
print(40 in nums)

# Dictionary
student = {"id": 101, "name": "Alice"}
print("id" in student)  #check only keys
print("Alice" in student)   # False (does NOT check values)

# If you want to check values:
print("Alice" in student.values())  # True
```

### 16] What is a nested dictionary?, and give an example of its use case.

Answer:- A nested dictionary is a dictionary inside a dictionary, useful for representing hierarchical or structured data, like student records or JSON responses.Useful when data has multiple levels of information.

### 19] Why are dictionaries considered unordered, and how does that affect data retrieval?

Answer:- Dictionaries are considered unordered because they're based on hash tables, meaning retrieval is key-based, not position-based. Since Python 3.7, insertion order is preserved, but you still can't access items by index like in lists.

* Effect on data retrieval:
1. You cannot rely on positional indexing like with lists (dict[0] is invalid).
2. Retrieval must always be done by key, not by position.
3. key-based lookup structure.

### 20] Explain the difference between a list and a dictionary in terms of data retrieval.

Answer:-

* list:-
      1. Data is retrieved by index (position).
      2. Example: my_list[0] → gets the first element.
      3. Time complexity: O(1) for direct index access, O(n) for searching a value.

* Dictionary:-
      1. Data is retrieved by key.
      2. Example: my_dict["id"] → gets the value for that key.
      3. Time complexity: O(1) on average due to hashing.
